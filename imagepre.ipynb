```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow

# Load the image
image = cv2.imread('/content/sample_data/cat.jpg')

# 1. Resize the image (to a specific width and height)
resized_image = cv2.resize(image, (640, 480))  # Resize to 640x480 pixels

# 2. Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 3. Apply Gaussian Blur (to reduce noise)
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# 4. Thresholding (Binary thresholding)
_, thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# 5. Edge Detection using Canny
edges = cv2.Canny(gray_image, 100, 200)

# 6. Dilation and Erosion
kernel = np.ones((5, 5), np.uint8)
dilated_image = cv2.dilate(thresholded_image, kernel, iterations=1)
eroded_image = cv2.erode(thresholded_image, kernel, iterations=1)

# 7. Histogram Equalization (Improves contrast)
equalized_image = cv2.equalizeHist(gray_image)

# Display the processed images using cv2_imshow
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
cv2_imshow(resized_image)
cv2_imshow(gray_image)
cv2_imshow(blurred_image)
cv2_imshow(thresholded_image)
cv2_imshow(edges)
cv2_imshow(dilated_image)
cv2_imshow(eroded_image)
cv2_imshow(equalized_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
